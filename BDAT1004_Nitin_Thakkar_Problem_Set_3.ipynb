<h1><font color='#004b87'>step 4</font></h1>

## Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
import datetime
def dateCorrection(date):
    if date.year > 1989:
        y = date.year - 100 
    else:
        y = date.year
    return datetime.date(y, date.month ,date.day)
```

```python
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(dateCorrection)
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

mrg_data = pd.read_csv("us-marriages-divorces-1867-2014.csv")
marriages_capita = mrg_data['Marriages_per_1000']
divorces_capita = mrg_data['Divorces_per_1000']

col = marriages_capita, divorces_capita
label_col = ["marriages per capita", "divorces per capita"]
clr=["blue","red"]

x_axis = range(1867, 1867 + mrg_data.shape[0])
fig, ax = plt.subplots()
# plt.figure(figsize=(10, 3))
plt.rcParams['figure.figsize'] = [2, 2]

for i in range(len(label_col)):
    ax.plot(x_axis, col[i], label=label_col[i],color=clr[i])

ax.set_xlabel('Year',fontsize=18)
ax.set_ylabel('Marriage or Divorces per 1000 people',fontsize=18)
ax.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

specific_data = data.loc[data['Year'] == 1900].append(data.loc[data['Year'] == 1950]).append(data.loc[data['Year'] == 2000])

plotdata = pd.DataFrame({
    "Marriages_per_1000":specific_data['Marriages_per_1000'].to_list(),
    "Divorces_per_1000":specific_data['Divorces_per_1000'].to_list()
    }, 
    index=[1900, 1950, 2000]
)
plt.rcParams["figure.figsize"] = [10, 6]
plotdata.plot(kind="bar")
plt.title("Marriage Vs Divorce By year")
plt.xlabel("Year",fontsize=18)
plt.ylabel("Marraige/ Divorced Per Capita",fontsize=18)
```
